Кількість позитивних чисел.

Добуток усіх негативних чисел.

Мінімальне та максимальне числа без використання функцій min() та max().


Виконайте приклад нижче:

```python
# Ввід рядка чисел
input_str = input("Введіть ціле числа через пробіл: ")

# Перетворення на список цілих чисел
numbers = list(map(int, input_str.split()))

# Ініціалізація змінних
positive_count = 0
negative_product = None  # None означає, що ще не було негативних чисел
min_num = numbers[0]
max_num = numbers[0]

# Обробка кожного числа
for num in numbers:
    if num > 0:
        positive_count += 1
    if num < 0:
        if negative_product is None:
            negative_product = num
        else:
            negative_product *= num
    # Пошук мінімального та максимального без min() і max()
    if num < min_num:
        min_num = num
    if num > max_num:
        max_num = num

# Вивід результатів
print("Кількість позитивних чисел:", positive_count)
print("Добуток негативних чисел:", negative_product if negative_product is not None else "Немає негативних чисел")
print("Мінімальне число:", min_num)
print("Максимальне число:",
```

**Завдання 6**

Напишіть програму, яка:

створює квадратну матрицю з отриманих на вхід рядків;

перевіряє, чи є матриця симетричною щодо побічної діагоналі;

виводить Так чи Ні залежно від результату.


Виконайте приклад нижче


```python
# Введення розміру матриці
n = int(input("Введіть розмір квадратної матриці: "))

# Введення матриці
print("Введіть матрицю по рядках (елементи через пробіл):")
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

# Перевірка симетрії щодо побічної діагоналі
is_symmetric = True
for i in range(n):
    for j in range(n):
        if matrix[i][j] != matrix[n - 1 - j][n - 1 - i]:
            is_symmetric = False
            break
    if not is_symmetric:
        break

# Вивід результату
print("Так" if is_symmetric else "Ні")
```
