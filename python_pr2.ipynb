**Завдання 3**

Напишіть програму, яка отримує рядок з ім'ям, по батькові та прізвищем, написаними у довільному регістрі, і виводить дані в правильному форматі. Наприклад, рядок "тарас гриГороВич шевЧенко" має бути перетворений на "Тарас Григорович Шевченко".

split() розбиває рядок на окремі слова.

capitalize() робить першу літеру великою, а решту — маленькими.

join() об’єднує слова назад у рядок з пробілами.

Виконайте приклад нижче

```python
# Ввід рядка
full_name = input("Введіть ім'я, по батькові та прізвище: ")

# Розділення на складові
parts = full_name.split()

# Перетворення кожної частини у правильний формат
formatted_parts = [part.capitalize() for part in parts]

# Об'єднання назад у рядок
formatted_name = ' '.join(formatted_parts)

# Вивід результату
print(formatted_name)
```

**Завдання 8**

Напишіть програму, яка приймає на вхід рядок тексту і підрядок, а потім виводить індекси першого входження підрядка з початку і з кінця рядка (без урахування регістру).

Приклад введення:

Шість шустрий мишат в очеретах шурхотять  
шу


```python
# Ввід тексту та підрядка
text = input("Введіть текст: ")
substring = input("Введіть підрядок: ")

# Перетворення обох рядків до нижнього регістру
text_lower = text.lower()
substring_lower = substring.lower()

# Індекси першого входження з початку і з кінця
first_index = text_lower.find(substring_lower)
last_index = text_lower.rfind(substring_lower)

# Вивід результату
print(first_index)
print(last_index)
```
