Виконайте приклад нижче:

```python
# Введення рядка
text = input("Введіть рядок: ")

# Створення списку лише з цифр
digits = [int(char) for char in text if char.isdigit()]

print("Мінімальна цифра:", min(digits))
print("Максимальна цифра:", max(digits))
print("Сума всіх цифр:", sum(digits))
```

**Завдання 4**

Напишіть програму, яка:

отримує на вхід рядок чисел, розділених комами;

формує з чисел два списки – з парними та непарними числами;

виводить списки без дужок на окремих рядках.


Виконайте приклад нижче:

```python
# Введення рядка чисел через кому
text = input("Введіть числа через кому: ")

# Перетворення на список цілих чисел
numbers = list(map(int, text.split(',')))

# Формування списків парних та непарних чисел
even = [num for num in numbers if num % 2 == 0]
odd = [num for num in numbers if num % 2 != 0]

# Виведення списків без дужок
print(*even)
print(*odd)
```

Напишіть програму, яка:

```python
#Введення рядка
text = input("Введіть рядок з символом #: ")

# Видалення символа # (лише першого входження)
index = text.index('#')
text = text[:index] + text[index+1:]

# Виведення виправленого рядка
print("Виправлений рядок:", text)

# Пошук найдовшого слова
words = text.split()
longest = max(words, key=len)

print("Найдовше слово:", longest)


```

**Завдання 6**

Напишіть програму, яка:

приймає на вхід рядок, що складається із випадкових слів;

відкидає слова, де є не літерні символи;

переставляє слова у порядку зростання їх довжини;

виводить на екран слова у стовпчик, без лапок.

Виконайте приклад нижче:


```python
# Введення рядка
text = input("Введіть рядок зі словами: ")

# Розбиття на слова
words = text.split()

# Відбір лише слів, що складаються з літер (без цифр, знаків тощо)
letter_words = [word for word in words if word.isalpha()]

# Сортування за довжиною
sorted_words = sorted(letter_words, key=len)

# Виведення у стовпчик
print(*sorted_words,sep="\n" )
```

**Завдання 7**

Напишіть програму, яка отримує на вхід рядок, та формує з неї 2 списки:

до першого списку входять слова, що починаються з голосних літер;

другий список складається зі слів, що починаються з приголосних.

Виведіть результат у два рядки без лапок та дужок.

Виконай приклад нижче:


```python
# Введення рядка
text = input("Введіть рядок зі словами: ")

# Список голосних літер (українських)
vowels = 'аеєиіїоуюяАЕЄИІЇОУЮЯ'

# Розбиття на слова
words = text.split()

# Розділення слів за першою літерою
vowel_words = [word for word in words if word[0] in vowels]
consonant_words = [word for word in words if word[0] not in vowels]

# Виведення результату без лапок і дужок
print(*vowel_words)
print(*consonant_words)
```

**Завдання 9**

Напишіть програму, яка отримує на вхід рядок, що містить ім'я учня та його оцінки з основних предметів, та виводить середній бал успішності.

```python
# Введення рядка
text = input("Введіть ім'я учня та його оцінки: ")

# Розбиття на частини
parts = text.split()

# Ім’я — перший елемент
name = parts[0]

# Оцінки — решта елементів, перетворені на числа
grades = list(map(float, parts[1:]))

# Обчислення середнього балу
average = sum(grades) / len(grades) if grades else 0

# Виведення результату
print(f"{name}: середній бал {average:.2f}")
```

**Завдання 10**

Іграшковий робот пересувається відповідно до команд, одержуваних у вигляді рядка виду n35 s48 e88 w33 n12 s10, де n – північ, s – південь, w – захід, e – схід, а цифри – відстань у сантиметрах. Напишіть програму для визначення x та y координат робота. Вважатимемо, що рух завжди починається з 0, 0.

```python
# Введення команд
commands = input("Введіть команди руху робота: ")

# Розбиття на окремі команди
steps = commands.split()

# Обчислення координат без циклів
x = sum([int(i[1:]) for i in steps if i[0] == 'e']) - sum([int(i[1:]) for i in steps if i[0] == 'w'])
y = sum([int(i[1:]) for i in steps if i[0] == 'n']) - sum([int(i[1:]) for i in steps if i[0] == 's'])

# Виведення результату
print(f"x = {x}, y = {y}")
```
