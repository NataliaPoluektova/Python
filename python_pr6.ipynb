**–ó–∞–≤–¥–∞–Ω–Ω—è 1**

–ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é, —è–∫–∞ –ø—Ä–∏–π–º–∞—î –¥–æ–≤—ñ–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ü—ñ–ª–∏—Ö —á–∏—Å–µ–ª, —ñ –ø–æ–≤–µ—Ä—Ç–∞—î —Å–µ—Ä–µ–¥–Ω—î –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–µ –±–µ–∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –≤–±—É–¥–æ–≤–∞–Ω–∏—Ö —Ñ—É–Ω–∫—Ü—ñ–π sum() —Ç–∞ len().


–í–∏–∫–æ–Ω–∞–π—Ç–µ –ø—Ä–∏–∫–ª–∞–¥ –Ω–∏–∂—á–µ

```python
–ó–≤—ñ—Å–Ω–æ, –æ—Å—å —Ñ—É–Ω–∫—Ü—ñ—è –Ω–∞ Python, —è–∫–∞ –æ–±—á–∏—Å–ª—é—î —Å–µ—Ä–µ–¥–Ω—î –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–µ –¥–æ–≤—ñ–ª—å–Ω–æ—ó –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —Ü—ñ–ª–∏—Ö —á–∏—Å–µ–ª, –Ω–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≤–±—É–¥–æ–≤–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—ó sum() —Ç–∞ len().

üíª –§—É–Ω–∫—Ü—ñ—è –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–æ–≥–æ
–¶—è —Ñ—É–Ω–∫—Ü—ñ—è –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î —Å–∏–Ω—Ç–∞–∫—Å–∏—Å *args, —â–æ–± –ø—Ä–∏–π–Ω—è—Ç–∏ –¥–æ–≤—ñ–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–∑–∏—Ü—ñ–π–Ω–∏—Ö –∞—Ä–≥—É–º–µ–Ω—Ç—ñ–≤, —ñ –æ–±—á–∏—Å–ª—é—î —Å—É–º—É —Ç–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –≤—Ä—É—á–Ω—É.

Python

def average_manual(*numbers):

    # –Ü–Ω—ñ—Ü—ñ–∞–ª—ñ–∑—É—î–º–æ –ª—ñ—á–∏–ª—å–Ω–∏–∫ —Ç–∞ –∑–∞–≥–∞–ª—å–Ω—É —Å—É–º—É
    total_sum = 0
    count = 0

    # 1. –û–±—á–∏—Å–ª–µ–Ω–Ω—è —Å—É–º–∏ —Ç–∞ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
    for number in numbers:
        # –î–æ–¥–∞—î–º–æ —á–∏—Å–ª–æ –¥–æ –∑–∞–≥–∞–ª—å–Ω–æ—ó —Å—É–º–∏
        total_sum += number
        # –ó–±—ñ–ª—å—à—É—î–º–æ –ª—ñ—á–∏–ª—å–Ω–∏–∫ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
        count += 1

    # 2. –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –Ω–∞ –≤–∏–ø–∞–¥–æ–∫, —è–∫—â–æ –∞—Ä–≥—É–º–µ–Ω—Ç–∏ –Ω–µ –ø–µ—Ä–µ–¥–∞–Ω–æ
    if count == 0:
        return 0  # –ü–æ–≤–µ—Ä—Ç–∞—î–º–æ 0, —è–∫—â–æ —Å–ø–∏—Å–æ–∫ –ø–æ—Ä–æ–∂–Ω—ñ–π, —â–æ–± —É–Ω–∏–∫–Ω—É—Ç–∏ ZeroDivisionError

    # 3. –û–±—á–∏—Å–ª–µ–Ω–Ω—è —Ç–∞ –ø–æ–≤–µ—Ä–Ω–µ–Ω–Ω—è —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–æ–≥–æ
    # –°–µ—Ä–µ–¥–Ω—î = –°—É–º–∞ / –ö—ñ–ª—å–∫—ñ—Å—Ç—å
    return total_sum / count

# –ü—Ä–∏–∫–ª–∞–¥ 1: –ó–≤–∏—á–∞–π–Ω–∏–π –Ω–∞–±—ñ—Ä —á–∏—Å–µ–ª
result_1 = average_manual(10, 20, 30, 40)
print(f"–°–µ—Ä–µ–¥–Ω—î –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–µ (10, 20, 30, 40): {result_1}")  # –í–∏–≤–µ–¥–µ: 25.0

# –ü—Ä–∏–∫–ª–∞–¥ 2: –ó –≤—ñ–¥'—î–º–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏
result_2 = average_manual(5, -5, 15)
print(f"–°–µ—Ä–µ–¥–Ω—î –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–µ (5, -5, 15): {result_2}")      # –í–∏–≤–µ–¥–µ: 5.0

# –ü—Ä–∏–∫–ª–∞–¥ 3: –ü—É—Å—Ç–∏–π –Ω–∞–±—ñ—Ä (–ø–æ–≤–µ—Ä–Ω–µ 0)
result_3 = average_manual()
print(f"–°–µ—Ä–µ–¥–Ω—î –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–µ (): {result_3}")              # –í–∏–≤–µ–¥–µ: 0
```

**–ó–∞–≤–¥–∞–Ω–Ω—è –¥–ª—è —Å–∞–º–æ—Å—Ç—ñ–π–Ω–æ–≥–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è 1**

–ó–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ—ó –Ω–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –≤–∏–≤–æ–¥–∏—Ç—å —Ç—ñ–ª—å–∫–∏ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ –∑—ñ —Å–ø–∏—Å–∫—É.

**–ó–∞–≤–¥–∞–Ω–Ω—è 2**

–ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —â–æ–± –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ –Ω–∞–¥—ñ–π–Ω—ñ—Å—Ç—å –ø–∞—Ä–æ–ª—è. –ù–∞–¥—ñ–π–Ω–∏–π –ø–∞—Ä–æ–ª—å –º–∞—î –º—ñ—Å—Ç–∏—Ç–∏ —â–æ–Ω–∞–π–º–µ–Ω—à–µ 9 —Å–∏–º–≤–æ–ª—ñ–≤; –∫—Ä—ñ–º —Ç–æ–≥–æ, —É –ø–∞—Ä–æ–ª—ñ –º–∞—î –±—É—Ç–∏ —Ö–æ—á–∞ –± –æ–¥–Ω–∞:
–ª—ñ—Ç–µ—Ä–∞ —É –≤–µ—Ä—Ö–Ω—å–æ–º—É —Ä–µ–≥—ñ—Å—Ç—Ä—ñ;
–ª—ñ—Ç–µ—Ä–∞ —É –Ω–∏–∂–Ω—å–æ–º—É —Ä–µ–≥—ñ—Å—Ç—Ä—ñ;
—Ü–∏—Ñ—Ä–∞;
—ñ —Ö–æ—á–∞ –± –æ–¥–∏–Ω —Å–∏–º–≤–æ–ª —ñ–∑ –Ω–∞–±–æ—Ä—É !@#$%^&*()-+.


–í–∏–∫–æ–Ω–∞–π—Ç–µ –ø—Ä–∏–∫–ª–∞–¥ –Ω–∏–∂—á–µ:

```python
def is_strong_password(password):
    if len(password) < 9:
        return False

    has_upper = any(c.isupper() for c in password)
    has_lower = any(c.islower() for c in password)
    has_digit = any(c.isdigit() for c in password)
    has_special = any(c in "!@#$%^&*()-+" for c in password)

    return has_upper and has_lower and has_digit and has_special

# –¢–µ—Å—Ç—É–≤–∞–Ω–Ω—è:
pw = input("–í–≤–µ–¥—ñ—Ç—å –ø–∞—Ä–æ–ª—å: ")
if is_strong_password(pw):
    print(" –ù–∞–¥—ñ–π–Ω–∏–π –ø–∞—Ä–æ–ª—å!")
else:
    print(" –ü–∞—Ä–æ–ª—å –Ω–µ–Ω–∞–¥—ñ–π–Ω–∏–π. –°–ø—Ä–æ–±—É–π—Ç–µ —â–µ —Ä–∞–∑.")

```

–ó–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—ó –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –æ—Ç—Ä–∏–º—É—î –≤—ñ–¥ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞ —Å–ø–∏—Å–æ–∫, —Ç–∞ –≤–∏–≤–æ–¥–∏—Ç—å –≤ –∞–ª—Ñ–∞–≤—ñ—Ç–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É –≤—Å—ñ —Å–ª–æ–≤–∞, —â–æ —Å–∫–ª–∞–¥–∞—é—Ç—å—Å—è –∑ 5 –ª—ñ—Ç–µ—Ä.


–í–∏–∫–æ–Ω–∞–π—Ç–µ –ø—Ä–∏–∫–ª–∞–¥ –Ω–∏–∂—á–µ:

```python
# 1. –û—Ç—Ä–∏–º–∞–Ω–Ω—è –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö –≤—ñ–¥ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞
print("–í–≤–µ–¥—ñ—Ç—å —Å–ª–æ–≤–∞, —Ä–æ–∑–¥—ñ–ª–µ–Ω—ñ –∫–æ–º–æ—é –∞–±–æ –ø—Ä–æ–±—ñ–ª–æ–º:")
user_input = input()

# 2. –û–±—Ä–æ–±–∫–∞ –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö: —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É —Å–ª—ñ–≤
# –†–æ–∑–±–∏–≤–∞—î–º–æ —Ä—è–¥–æ–∫ –∑–∞ –ø—Ä–æ–±—ñ–ª–∞–º–∏ –∞–±–æ –∫–æ–º–∞–º–∏ —Ç–∞ –≤–∏–¥–∞–ª—è—î–º–æ –∑–∞–π–≤—ñ –ø—Ä–æ–±—ñ–ª–∏ –Ω–∞–≤–∫–æ–ª–æ —Å–ª—ñ–≤
words = [word.strip() for word in user_input.replace(',', ' ').split() if word.strip()]

# 3. –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Ç–∞ –∑–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—ó –¥–ª—è —Ñ—ñ–ª—å—Ç—Ä–∞—Ü—ñ—ó
# –§—ñ–ª—å—Ç—Ä—É—î–º–æ —Å–ª–æ–≤–∞, –∑–∞–ª–∏—à–∞—é—á–∏ –ª–∏—à–µ —Ç—ñ, –¥–æ–≤–∂–∏–Ω–∞ —è–∫–∏—Ö –¥–æ—Ä—ñ–≤–Ω—é—î 5
filter_5_letters = lambda w: len(w) == 5

# –ó–∞—Å—Ç–æ—Å–æ–≤—É—î–º–æ —Ñ—ñ–ª—å—Ç—Ä
filtered_words = list(filter(filter_5_letters, words))

# 4. –°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É –≤ –∞–ª—Ñ–∞–≤—ñ—Ç–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É
# –°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Å–ø–∏—Å–∫—É –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –≤–±—É–¥–æ–≤–∞–Ω–∏–º –º–µ—Ç–æ–¥–æ–º sort()
filtered_words.sort()

# 5. –í–∏–≤–µ–¥–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É
if filtered_words:
    print("\n–°–ª–æ–≤–∞ –∑ 5 –ª—ñ—Ç–µ—Ä (–≤ –∞–ª—Ñ–∞–≤—ñ—Ç–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É):")
    print(filtered_words)
else:
    print("\n–£ –≤–≤–µ–¥–µ–Ω–æ–º—É —Å–ø–∏—Å–∫—É –Ω–µ–º–∞—î —Å–ª—ñ–≤, —â–æ —Å–∫–ª–∞–¥–∞—é—Ç—å—Å—è –∑ 5 –ª—ñ—Ç–µ—Ä.")
```

–ó–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—ó –Ω–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –≤–∏–≤–æ–¥–∏—Ç—å —Ç—ñ–ª—å–∫–∏ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ –∑—ñ —Å–ø–∏—Å–∫—É

```python
#–í–∞—à –∫–æ–¥ —Ç—É—Ç:
```

**–Ü–Ω–¥–∏–≤—ñ–¥—É–∞–ª—å–Ω–µ –∑–∞–≤–¥–∞–Ω–Ω—è 2 –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—ó**

–í–∏–∫–æ–Ω–∞–π—Ç–µ –æ–¥–Ω—É –∑–∞–¥–∞—á—É —Å–≤–æ–≥–æ –≤–∞—Ä—ñ–∞–Ω—Ç—É


1.	–°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Å–ø–∏—Å–∫—É —Å–ª—ñ–≤ –∑–∞ –¥–æ–≤–∂–∏–Ω–æ—é: –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Å–ø–∏—Å–∫—É —Å–ª—ñ–≤ –∑–∞ —ó—Ö –¥–æ–≤–∂–∏–Ω–æ—é.

2.	–§—ñ–ª—å—Ç—Ä—É–≤–∞–Ω–Ω—è –ø–∞—Ä–Ω–∏—Ö —á–∏—Å–µ–ª: –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è —Ñ—ñ–ª—å—Ç—Ä–∞—Ü—ñ—ó –ª–∏—à–µ –ø–∞—Ä–Ω–∏—Ö —á–∏—Å–µ–ª –∑—ñ —Å–ø–∏—Å–∫—É.

3.	–ü—ñ–¥—Ä–∞—Ö—É–Ω–æ–∫ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –±—É–∫–≤ 'a' —É —Ä—è–¥–∫—É: –ù–∞–ø–∏—à—ñ—Ç—å –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é, —è–∫–∞ –≤–≤–∞–∂–∞—î –∫—ñ–ª—å–∫—ñ—Å—Ç—å –±—É–∫–≤ 'a' —É —Ä—è–¥–∫—É.

4.	–°—É–º–∞ –∫–≤–∞–¥—Ä–∞—Ç—ñ–≤: –ù–∞–ø–∏—à—ñ—Ç—å –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é, —è–∫–∞ –ø—Ä–∏–π–º–∞—î –¥–≤–∞ —á–∏—Å–ª–∞ —Ç–∞ –ø–æ–≤–µ—Ä—Ç–∞—î —Å—É–º—É —ó—Ö –∫–≤–∞–¥—Ä–∞—Ç—ñ–≤.

5.	–§—ñ–ª—å—Ç—Ä—É–≤–∞–Ω–Ω—è —Ä—è–¥–∫—ñ–≤, —â–æ –ø–æ—á–∏–Ω–∞—é—Ç—å—Å—è –∑ –ø–µ–≤–Ω–æ—ó –ª—ñ—Ç–µ—Ä–∏: –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è —Ñ—ñ–ª—å—Ç—Ä–∞—Ü—ñ—ó —Ä—è–¥–∫—ñ–≤, —è–∫—ñ –ø–æ—á–∏–Ω–∞—é—Ç—å—Å—è –∑ –ª—ñ—Ç–µ—Ä–∏ 'a'.

6.	–°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Å–ø–∏—Å–∫—É —Å–ª—ñ–≤ –∑–∞ –æ—Å—Ç–∞–Ω–Ω—å–æ—é –ª—ñ—Ç–µ—Ä–æ—é: –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Å–ø–∏—Å–∫—É —Å–ª—ñ–≤ –∑–∞ –æ—Å—Ç–∞–Ω–Ω—å–æ—é –ª—ñ—Ç–µ—Ä–æ—é.

7.	–ü—ñ–¥—Ä–∞—Ö—É–Ω–æ–∫ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —Å–ª—ñ–≤ —É —Ä—è–¥–∫—É: –ù–∞–ø–∏—à—ñ—Ç—å –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é, —è–∫–∞ –ø–æ–≤–µ—Ä—Ç–∞—î –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Å–ª—ñ–≤ —É —Ä—è–¥–∫—É.

8.	–û–±'—î–¥–Ω–∞–Ω–Ω—è —Å–ø–∏—Å–∫—É —Ä—è–¥–∫—ñ–≤ —ñ–∑ –ø—Ä–æ–±—ñ–ª–æ–º: –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è –æ–±'—î–¥–Ω–∞–Ω–Ω—è —Å–ø–∏—Å–∫—É —Ä—è–¥–∫—ñ–≤ –≤ –æ–¥–∏–Ω —Ä—è–¥–æ–∫ —ñ–∑ –ø—Ä–æ–±—ñ–ª–∞–º–∏ –º—ñ–∂ —Å–ª–æ–≤–∞–º–∏.

9.	–§—ñ–ª—å—Ç—Ä—É–≤–∞–Ω–Ω—è —á–∏—Å–µ–ª, –∫—Ä–∞—Ç–Ω–∏—Ö 3 —Ç–∞ 5: –ù–∞–ø–∏—à—ñ—Ç—å –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è —Ñ—ñ–ª—å—Ç—Ä–∞—Ü—ñ—ó —á–∏—Å–µ–ª, –∫—Ä–∞—Ç–Ω–∏—Ö 3 —Ç–∞ 5 –∑—ñ —Å–ø–∏—Å–∫—É.

10.	–ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É —á–∏—Å–µ–ª –Ω–∞ —ó—Ö–Ω—ñ –∫–≤–∞–¥—Ä–∞—Ç–∏: –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É —á–∏—Å–µ–ª –Ω–∞ —ó—Ö–Ω—ñ –∫–≤–∞–¥—Ä–∞—Ç–∏.

11.	–°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Å–ø–∏—Å–∫—É –∫–æ—Ä—Ç–µ–∂—ñ–≤ –∑–∞ –¥—Ä—É–≥–∏–º –µ–ª–µ–º–µ–Ω—Ç–æ–º: –ù–∞–ø–∏—à—ñ—Ç—å –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Å–ø–∏—Å–∫—É –∫–æ—Ä—Ç–µ–∂—ñ–≤ –∑–∞ –¥—Ä—É–≥–∏–º –µ–ª–µ–º–µ–Ω—Ç–æ–º.

12.	–ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É —Ä—è–¥–∫—ñ–≤ –Ω–∞ —Å–ø–∏—Å–æ–∫ —ó—Ö –¥–æ–≤–∂–∏–Ω: –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É —Ä—è–¥–∫—ñ–≤ –Ω–∞ —Å–ø–∏—Å–æ–∫ —ó—Ö –¥–æ–≤–∂–∏–Ω.

13.	–§—ñ–ª—å—Ç—Ä—É–≤–∞–Ω–Ω—è –ø–æ–∑–∏—Ç–∏–≤–Ω–∏—Ö —á–∏—Å–µ–ª: –ù–∞–ø–∏—à—ñ—Ç—å –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è —Ñ—ñ–ª—å—Ç—Ä–∞—Ü—ñ—ó –ø–æ–∑–∏—Ç–∏–≤–Ω–∏—Ö —á–∏—Å–µ–ª –∑—ñ —Å–ø–∏—Å–∫—É.

14.	–°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Å–ª—ñ–≤ –∑–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—é –≥–æ–ª–æ—Å–Ω–∏—Ö: –ù–∞–ø–∏—à—ñ—Ç—å –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Å–ø–∏—Å–∫—É —Å–ª—ñ–≤ –∑–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—é –≥–æ–ª–æ—Å–Ω–∏—Ö —É –∫–æ–∂–Ω–æ–º—É —Å–ª–æ–≤—ñ.

15.	–ü—ñ–¥—Ä–∞—Ö—É–Ω–æ–∫ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —á–∏—Å–µ–ª –±—ñ–ª—å—à–µ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ: –ù–∞–ø–∏—à—ñ—Ç—å –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é, —è–∫–∞ –≤–≤–∞–∂–∞—î –∫—ñ–ª—å–∫—ñ—Å—Ç—å —á–∏—Å–µ–ª —É —Å–ø–∏—Å–∫—É, —è–∫—ñ –±—ñ–ª—å—à—ñ –∑–∞ —Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É.



```python
# –í–∞—à –∫–æ–¥ —Ç—É—Ç:
```

**–ó–∞–≤–¥–∞–Ω–Ω—è –¥–ª—è —Å–∞–º–æ—Å—Ç—ñ–π–Ω–æ–≥–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è 4**

–£ –≤–∞—Å —î —Å–ø–∏—Å–æ–∫, –¥–µ –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç ‚Äì —Ü–µ —Å–ª–æ–≤–Ω–∏–∫, —â–æ –º—ñ—Å—Ç–∏—Ç—å —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—é –ø—Ä–æ —Ç–æ–≤–∞—Ä ('name' —Ç–∞ 'price'). –í—ñ–¥—Å–æ—Ä—Ç—É–π—Ç–µ —Ü–µ–π —Å–ø–∏—Å–æ–∫ –∑–∞ —Ü—ñ–Ω–æ—é (–∑–Ω–∞—á–µ–Ω–Ω—è –∫–ª—é—á–∞ 'price').

–ü–æ—è—Å–Ω–µ–Ω–Ω—è
–°–ø–∏—Å–æ–∫ —Å–ª–æ–≤–Ω–∏–∫—ñ–≤ (products): –¶–µ –æ—Å–Ω–æ–≤–Ω–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–∞–Ω–∏—Ö, —è–∫—É –º–∏ —Å–æ—Ä—Ç—É—î–º–æ. –ö–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç —Å–ø–∏—Å–∫—É —î —Å–ª–æ–≤–Ω–∏–∫–æ–º.

–§—É–Ω–∫—Ü—ñ—è –≤–∏—â–æ–≥–æ –ø–æ—Ä—è–¥–∫—É sorted(): –í–æ–Ω–∞ –ø—Ä–∏–π–º–∞—î —Å–ø–∏—Å–æ–∫ products —ñ —Ñ—É–Ω–∫—Ü—ñ—é-–∞—Ä–≥—É–º–µ–Ω—Ç key.

–õ—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—è (lambda item: item['price']):

–í–æ–Ω–∞ –ø—Ä–∏–π–º–∞—î –æ–¥–∏–Ω –∞—Ä–≥—É–º–µ–Ω—Ç item (—è–∫–∏–π —î –æ–¥–Ω–∏–º —Å–ª–æ–≤–Ω–∏–∫–æ–º –∑—ñ —Å–ø–∏—Å–∫—É).

–ü–æ–≤–µ—Ä—Ç–∞—î –∑–Ω–∞—á–µ–Ω–Ω—è, –ø–æ–≤'—è–∑–∞–Ω–µ –∑ –∫–ª—é—á–µ–º 'price' —É —Ü—å–æ–º—É —Å–ª–æ–≤–Ω–∏–∫—É.

–§—É–Ω–∫—Ü—ñ—è sorted() –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î —Ü–µ —Ü—ñ–Ω–æ–≤–µ –∑–Ω–∞—á–µ–Ω–Ω—è –¥–ª—è –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è —Ç–∞ –≤–ø–æ—Ä—è–¥–∫—É–≤–∞–Ω–Ω—è —Å–ª–æ–≤–Ω–∏–∫—ñ–≤ —É —Ñ—ñ–Ω–∞–ª—å–Ω–æ–º—É —Å–ø–∏—Å–∫—É.

–¢–∞–∫–∏–º —á–∏–Ω–æ–º, —Ñ—É–Ω–∫—Ü—ñ—è sorted() —î —Ñ—É–Ω–∫—Ü—ñ—î—é –≤–∏—â–æ–≥–æ –ø–æ—Ä—è–¥–∫—É, –æ—Å–∫—ñ–ª—å–∫–∏ –≤–æ–Ω–∞ –ø—Ä–∏–π–º–∞—î —ñ–Ω—à—É —Ñ—É–Ω–∫—Ü—ñ—é (lambda) —è–∫ –∞—Ä–≥—É–º–µ–Ω—Ç.

–í–∏–∫–æ–Ω–∞–π—Ç–µ –ø—Ä–∏–∫–ª–∞–¥ –Ω–∏–∂—á–µ

```python
products = [
    {'name': '–ù–æ—É—Ç–±—É–∫', 'price': 1200},
    {'name': '–ú–∏—à–∫–∞', 'price': 35},
    {'name': '–ú–æ–Ω—ñ—Ç–æ—Ä', 'price': 350},
    {'name': '–ö–ª–∞–≤—ñ–∞—Ç—É—Ä–∞', 'price': 80},
    {'name': '–í–µ–±–∫–∞–º–µ—Ä–∞', 'price': 350} # –¢–∞ —Å–∞–º–∞ —Ü—ñ–Ω–∞, —â–æ –π —É –ú–æ–Ω—ñ—Ç–æ—Ä–∞
]

# –§—É–Ω–∫—Ü—ñ—è –≤–∏—â–æ–≥–æ –ø–æ—Ä—è–¥–∫—É: sorted()
# –§—É–Ω–∫—Ü—ñ—è-–∞—Ä–≥—É–º–µ–Ω—Ç (callback): lambda item: item['price']

# –õ—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—è –≤–∫–∞–∑—É—î sorted(), —â–æ –¥–ª—è —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –ø–æ—Ç—Ä—ñ–±–Ω–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏
# –∑–Ω–∞—á–µ–Ω–Ω—è, –ø–æ–≤'—è–∑–∞–Ω–µ –∑ –∫–ª—é—á–µ–º 'price' —É –∫–æ–∂–Ω–æ–º—É —Å–ª–æ–≤–Ω–∏–∫—É (item).
sorted_by_price = sorted(products, key=lambda item: item['price'])

print("–í—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π —Å–ø–∏—Å–æ–∫ —Ç–æ–≤–∞—Ä—ñ–≤ –∑–∞ –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è–º —Ü—ñ–Ω–∏:")
for product in sorted_by_price:
    print(f"–¢–æ–≤–∞—Ä: {product['name']}, –¶—ñ–Ω–∞: {product['price']} $")
```

** –ó–∞–≤–¥–∞–Ω–Ω—è –¥–ª—è —Å–∞–º–æ—Å—Ç—ñ–π–Ω–æ–≥–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è 4**
