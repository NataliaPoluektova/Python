Напишіть функцію, яка приймає довільну кількість цілих чисел, і повертає середнє арифметичне без використання вбудованих функцій sum() та len().


Виконайте приклад нижче

```python
Звісно, ось функція на Python, яка обчислює середнє арифметичне довільної кількості цілих чисел, не використовуючи вбудовані функції sum() та len().

💻 Функція для обчислення середнього арифметичного
Ця функція використовує синтаксис *args, щоб прийняти довільну кількість позиційних аргументів, і обчислює суму та кількість елементів вручну.

Python

def average_manual(*numbers):

    # Ініціалізуємо лічильник та загальну суму
    total_sum = 0
    count = 0

    # 1. Обчислення суми та кількості елементів
    for number in numbers:
        # Додаємо число до загальної суми
        total_sum += number
        # Збільшуємо лічильник елементів
        count += 1

    # 2. Перевірка на випадок, якщо аргументи не передано
    if count == 0:
        return 0  # Повертаємо 0, якщо список порожній, щоб уникнути ZeroDivisionError

    # 3. Обчислення та повернення середнього арифметичного
    # Середнє = Сума / Кількість
    return total_sum / count

# Приклад 1: Звичайний набір чисел
result_1 = average_manual(10, 20, 30, 40)
print(f"Середнє арифметичне (10, 20, 30, 40): {result_1}")  # Виведе: 25.0

# Приклад 2: З від'ємними числами
result_2 = average_manual(5, -5, 15)
print(f"Середнє арифметичне (5, -5, 15): {result_2}")      # Виведе: 5.0

# Приклад 3: Пустий набір (поверне 0)
result_3 = average_manual()
print(f"Середнє арифметичне (): {result_3}")              # Виведе: 0
```

**Завдання 2**

Напишіть програму, щоб перевірити надійність пароля. Надійний пароль має містити щонайменше 9 символів; крім того, у паролі має бути хоча б одна:
літера у верхньому регістрі;
літера у нижньому регістрі;
цифра;
і хоча б один символ із набору !@#$%^&*()-+.


Виконайте приклад нижче:

```python
def is_strong_password(password):
    if len(password) < 9:
        return False

    has_upper = any(c.isupper() for c in password)
    has_lower = any(c.islower() for c in password)
    has_digit = any(c.isdigit() for c in password)
    has_special = any(c in "!@#$%^&*()-+" for c in password)

    return has_upper and has_lower and has_digit and has_special

# Тестування:
pw = input("Введіть пароль: ")
if is_strong_password(pw):
    print(" Надійний пароль!")
else:
    print(" Пароль ненадійний. Спробуйте ще раз.")

```

**Завдання 3**
За допомогою лямбда-функції Напишіть програму, яка отримує від користувача список, та виводить в алфавітному порядку всі слова, що складаються з 5 літер.


Виконайте приклад нижче:

```python
# 1. Отримання вхідних даних від користувача
print("Введіть слова, розділені комою або пробілом:")
user_input = input()

# 2. Обробка вхідних даних: створення списку слів
# Розбиваємо рядок за пробілами або комами та видаляємо зайві пробіли навколо слів
words = [word.strip() for word in user_input.replace(',', ' ').split() if word.strip()]

# 3. Визначення та застосування лямбда-функції для фільтрації
# Фільтруємо слова, залишаючи лише ті, довжина яких дорівнює 5
filter_5_letters = lambda w: len(w) == 5

# Застосовуємо фільтр
filtered_words = list(filter(filter_5_letters, words))

# 4. Сортування результату в алфавітному порядку
# Сортування списку виконується вбудованим методом sort()
filtered_words.sort()

# 5. Виведення результату
if filtered_words:
    print("\nСлова з 5 літер (в алфавітному порядку):")
    print(filtered_words)
else:
    print("\nУ введеному списку немає слів, що складаються з 5 літер.")
```

**Завдання 4**

У вас є список, де кожен елемент – це словник, що містить інформацію про товар ('name' та 'price'). Відсортуйте цей список за ціною (значення ключа 'price').

Пояснення
Список словників (products): Це основна структура даних, яку ми сортуємо. Кожен елемент списку є словником.

Функція вищого порядку sorted(): Вона приймає список products і функцію-аргумент key.

Лямбда-функція (lambda item: item['price']):

Вона приймає один аргумент item (який є одним словником зі списку).

Повертає значення, пов'язане з ключем 'price' у цьому словнику.

Функція sorted() використовує це цінове значення для порівняння та впорядкування словників у фінальному списку.

Таким чином, функція sorted() є функцією вищого порядку, оскільки вона приймає іншу функцію (lambda) як аргумент.

Виконайте приклад нижче

```python
products = [
    {'name': 'Ноутбук', 'price': 1200},
    {'name': 'Мишка', 'price': 35},
    {'name': 'Монітор', 'price': 350},
    {'name': 'Клавіатура', 'price': 80},
    {'name': 'Вебкамера', 'price': 350} # Та сама ціна, що й у Монітора
]

# Функція вищого порядку: sorted()
# Функція-аргумент (callback): lambda item: item['price']

# Лямбда-функція вказує sorted(), що для сортування потрібно використовувати
# значення, пов'язане з ключем 'price' у кожному словнику (item).
sorted_by_price = sorted(products, key=lambda item: item['price'])

print("Відсортований список товарів за зростанням ціни:")
for product in sorted_by_price:
    print(f"Товар: {product['name']}, Ціна: {product['price']} $")
```
