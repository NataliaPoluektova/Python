Напишіть клас MusicAlbum, який має:
Атрибути title, artist, release_year, genre, tracklist (назва, виконавець, рік виходу, жанр, список треків).
Метод play_random_track() для виведення випадкової назви пісні.



Виконайте приклад нижче

```python
import random

class MusicAlbum:
    """
    Представляє музичний альбом з його основними характеристиками
    та можливістю відтворення випадкового треку.
    """

    def __init__(self, title, artist, release_year, genre, tracklist):
        """
        Конструктор для ініціалізації атрибутів альбому.

        :param title: Назва альбому (str).
        :param artist: Виконавець альбому (str).
        :param release_year: Рік виходу альбому (int).
        :param genre: Жанр альбому (str).
        :param tracklist: Список треків альбому (list of str).
        """
        self.title = title
        self.artist = artist
        self.release_year = release_year
        self.genre = genre
        self.tracklist = tracklist

    def play_random_track(self):
        """
        Вибирає та виводить випадкову назву пісні зі списку треків (tracklist).
        """
        # Перевірка, чи список треків не порожній
        if not self.tracklist:
            print(f"Альбом '{self.title}' не містить треків.")
            return

        # Використання функції random.choice() для вибору випадкового елемента зі списку
        random_track = random.choice(self.tracklist)

        print("Зараз грає випадковий трек:")
        print(f"Альбом: {self.title} - Виконавець: {self.artist}")
        print(f"Трек: **{random_track}**")

# ----------------- Приклад використання класу -----------------

# Створення екземпляра класу MusicAlbum
my_album = MusicAlbum(
    title="The Dark Side of the Moon",
    artist="Pink Floyd",
    release_year=1973,
    genre="Progressive Rock",
    tracklist=[
        "Speak to Me",
        "Breathe (In the Air)",
        "On the Run",
        "Time",
        "The Great Gig in the Sky",
        "Money",
        "Us and Them",
        "Any Colour You Like",
        "Brain Damage",
        "Eclipse"
    ]
)

print(f"--- Інформація про альбом ---")
print(f"Назва: {my_album.title}")
print(f"Виконавець: {my_album.artist}")
print(f"Треків у списку: {len(my_album.tracklist)}")
print("-----------------------------\n")

# Виклик методу для відтворення випадкового треку
my_album.play_random_track()

print("\n")

# Виклик методу ще раз для демонстрації випадковості
my_album.play_random_track()

```

**Завдання 2**

Напишіть суперклас Publisher (видавництво) та два підкласи BookPublisher (книжкове видавництво) та NewspaperPublisher (газетне видавництво).

Батьківський клас Publisher має два атрибути name і location (назва, розташування) та два методи:

get_info(self) – надає інформацію про назву та розташування видавництва;

publish(self, message) – виводить інформацію про видання, що міститься у друку.

Підкласи BookPublisher і NewspaperPublisher використовують метод super().
__init__(name, location) суперкласу для виведення інформації про свої назви та розташування, і крім того, мають власні атрибути:

BookPublisher - num_authors (кількість авторів).

NewspaperPublisher - num_pages (кількість сторінок в газеті).


Виконайте приклад нижче:

```python
class Publisher:
    """
    Суперклас, що представляє видавництво.
    """
    def __init__(self, name, location):
        """
        Конструктор для ініціалізації назви та розташування видавництва.
        """
        self.name = name
        self.location = location
        print(f"Створено видавництво: {self.name} ({self.location})")

    def get_info(self):
        """
        Повертає рядок з інформацією про назву та розташування видавництва.
        """
        return f"Видавництво: {self.name}, Розташування: {self.location}"

    def publish(self, message):
        """
        Виводить інформацію про видання.
        """
        print(f"[{self.name}] Приймає до друку повідомлення: '{message}'")


class BookPublisher(Publisher):
    """
    Підклас для книжкового видавництва. Додає атрибут num_authors.
    """
    def __init__(self, name, location, num_authors):
        # Виклик конструктора батьківського класу для ініціалізації name та location
        super().__init__(name, location)
        self.num_authors = num_authors

    def get_info(self):
        """
        Розширення методу get_info() батьківського класу.
        """
        # Використовуємо метод батьківського класу для базової інформації
        base_info = super().get_info()
        return f"{base_info}, Тип: Книжкове, Кількість авторів: {self.num_authors}"

    def publish(self, title):
        """
        Перевизначення методу publish() для контексту книги.
        """
        super().publish(f"Нова книга '{title}'")


class NewspaperPublisher(Publisher):
    """
    Підклас для газетного видавництва. Додає атрибут num_pages.
    """
    def __init__(self, name, location, num_pages):
        # Виклик конструктора батьківського класу для ініціалізації name та location
        super().__init__(name, location)
        self.num_pages = num_pages

    def get_info(self):
        """
        Розширення методу get_info() батьківського класу.
        """
        base_info = super().get_info()
        return f"{base_info}, Тип: Газетне, Кількість сторінок: {self.num_pages}"

    def publish(self, edition_date):
        """
        Перевизначення методу publish() для контексту газети.
        """
        super().publish(f"Газета за {edition_date} ({self.num_pages} сторінок)")

# Створення екземпляра книжкового видавництва
book_publisher = BookPublisher(
    name="Книжковий Клуб",
    location="Київ",
    num_authors=150
)

# Створення екземпляра газетного видавництва
newspaper_publisher = NewspaperPublisher(
    name="Щоденні Новини",
    location="Львів",
    num_pages=16
)

print("\n--- Інформація та публікації ---")

# Виклик методів BookPublisher
print(book_publisher.get_info())
book_publisher.publish("Майбутнє штучного інтелекту")

print("-" * 20)

# Виклик методів NewspaperPublisher
print(newspaper_publisher.get_info())
newspaper_publisher.publish("2025-11-17")
```
