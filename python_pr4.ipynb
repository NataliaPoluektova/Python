–í–∏–∫–æ–Ω–∞–π—Ç–µ –ø—Ä–∏–∫–ª–∞–¥ –Ω–∏–∂—á–µ

```python
# –í–≤—ñ–¥ –¥–≤–æ—Ö —Ä—è–¥–∫—ñ–≤
keys_line = input("–í–≤–µ–¥—ñ—Ç—å —Å–ª–æ–≤–∞ (—á–µ—Ä–µ–∑ –ø—Ä–æ–±—ñ–ª): ")
values_line = input("–í–≤–µ–¥—ñ—Ç—å —á–∏—Å–ª–∞ (—á–µ—Ä–µ–∑ –ø—Ä–æ–±—ñ–ª): ")

# –†–æ–∑–±–∏—Ç—Ç—è —Ä—è–¥–∫—ñ–≤ –Ω–∞ —Å–ø–∏—Å–∫–∏
keys = keys_line.split()
values = list(map(int, values_line.split()))

# –§–æ—Ä–º—É–≤–∞–Ω–Ω—è —Å–ª–æ–≤–Ω–∏–∫–∞
result = dict(zip(keys, values))

# –í–∏–≤—ñ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É
print(result)

```

–ü—Ä–æ–≥—Ä–∞–º–∞ –º–∞—î —Å—Ñ–æ—Ä–º—É–≤–∞—Ç–∏ —Å–ª–æ–≤–Ω–∏–∫, –¥–µ –∫–ª—é—á–∞–º–∏ —î –Ω–∞–∑–≤–∏ —Ç–æ–≤–∞—Ä—ñ–≤, –∞ –∑–Ω–∞—á–µ–Ω–Ω—è–º–∏ ‚Äî —ó—Ö–Ω—ñ —Ü—ñ–Ω–∏.

```python
text = input("–í–≤–µ–¥—ñ—Ç—å —Ç–µ–∫—Å—Ç: ").lower()

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–ª–æ–≤–Ω–∏–∫–∞ —á–µ—Ä–µ–∑ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä
char_count = {char: text.count(char) for char in set(text)}

# –í–∏–≤—ñ–¥ –±–µ–∑ —Ñ—ñ–≥—É—Ä–Ω–∏—Ö –¥—É–∂–æ–∫
for char, count in char_count.items():
    print(f"{char}: {count}")
```

–í–∏–≤–æ–¥–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç —É —Ñ–æ—Ä–º–∞—Ç—ñ —Ü–∏—Ñ—Ä–∞: –∫—ñ–ª—å–∫—ñ—Å—Ç—å

–ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –æ—Ç—Ä–∏–º—É—î 2 —Å–ª–æ–≤–∞ –Ω–∞ —Ä—ñ–∑–Ω–∏—Ö —Ä—è–¥–∫–∞—Ö —Ç–∞ –≤–∏–∑–Ω–∞—á–∞—î, —á–∏ —î —Ü—ñ —Å–ª–æ–≤–∞ –∞–Ω–∞–≥—Ä–∞–º–∞–º–∏ (–∑ –±—É–∫–≤ –æ–¥–Ω–æ–≥–æ —Å–ª–æ–≤–∞ –º–æ–∂–Ω–∞ —Å–∫–ª–∞—Å—Ç–∏ —ñ–Ω—à–µ).


–í–∏–∫–æ–Ω–∞–π—Ç–µ –ø—Ä–∏–∫–ª–∞–¥ –Ω–∏–∂—á–µ

```python
# –í–≤—ñ–¥ –¥–≤–æ—Ö —Å–ª—ñ–≤
word1 = input("–ü–µ—Ä—à–µ —Å–ª–æ–≤–æ: ").lower()
word2 = input("–î—Ä—É–≥–µ —Å–ª–æ–≤–æ: ").lower()

# –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ —á–µ—Ä–µ–∑ –º–Ω–æ–∂–∏–Ω–∏
is_anagram = set(word1) == set(word2)

# –í–∏–≤—ñ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É
print(is_anagram)
```

–û—Ç—Ä–∏–º—É—î –Ω–∞ –≤—Ö—ñ–¥ –¥–≤–∞ —Å–ª–æ–≤–∞ (–Ω–∞ –æ–∫—Ä–µ–º–∏—Ö —Ä—è–¥–∫–∞—Ö).

–°—Ç–≤–æ—Ä—é—î –º–Ω–æ–∂–∏–Ω–∏ –ª—ñ—Ç–µ—Ä –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ —Å–ª–æ–≤–∞.

**–ó–∞–≤–¥–∞–Ω–Ω—è 4**

–ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ —Ñ–æ—Ä–º—É—î —Ç–µ–ª–µ—Ñ–æ–Ω–Ω–∏–π –¥–æ–≤—ñ–¥–Ω–∏–∫ —ñ–∑ –≤–≤–µ–¥–µ–Ω–æ–≥–æ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–µ–º —Ä—è–¥–∫–∞, –∞ –ø–æ—Ç—ñ–º –≤–∏–¥–∞—î –ø–æ—Ç—Ä—ñ–±–Ω–∏–π –Ω–æ–º–µ—Ä –Ω–∞ —ñ–º'—è –∞–±–æ–Ω–µ–Ω—Ç–∞. –Ø–∫—â–æ –∞–±–æ–Ω–µ–Ω—Ç–∞ –∑ —Ç–∞–∫–æ—é –Ω–∞–∑–≤–æ—é –Ω–µ–º–∞—î, –ø–æ—Ç—Ä—ñ–±–Ω–æ –≤–∏–≤–µ—Å—Ç–∏ –ù–µ–º–∞—î —Ç–∞–∫–æ–≥–æ –∞–±–æ–Ω–µ–Ω—Ç–∞.
